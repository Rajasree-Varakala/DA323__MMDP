```python
import requests
from bs4 import BeautifulSoup
import re
import os
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# Function to extract text content from a webpage
def extract_text(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Remove all script and style elements
        for script in soup(["script", "style"]):
            script.decompose()
        
        # Get text from the BeautifulSoup object
        text = soup.get_text()
        
        # Clean and format text
        lines = (line.strip() for line in text.splitlines())
        chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
        text = '\n'.join(chunk for chunk in chunks if chunk)
        
        return text
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return ""

# List of 20 music-related categories with three websites each
categories = {
    "Musical Instruments": ["https://www.guitarworld.com", "https://www.pianostreet.com", "https://www.woodwind.org"],
    "Music Genres": ["https://www.allmusic.com", "https://rateyourmusic.com", "https://www.billboard.com"],
    "Music Theory": ["https://www.musictheory.net", "https://www.8notes.com", "https://www.musical-u.com"],
    "Music Production": ["https://www.soundonsound.com", "https://www.musicradar.com", "https://www.pro-tools-expert.com"],
    "Sheet Music": ["https://www.musicnotes.com", "https://www.sheetmusicplus.com", "https://www.8notes.com"],
    "Music History": ["https://www.rockhall.com", "https://www.classicfm.com", "https://www.allmusic.com"],
    "Concerts & Festivals": ["https://www.pollstar.com", "https://www.festicket.com", "https://www.rollingstone.com/music/music-news"],
    "Music Streaming Services": ["https://www.spotify.com", "https://www.apple.com/music", "https://www.tidal.com"],
    "Famous Musicians": ["https://www.biography.com", "https://www.billboard.com", "https://www.nme.com"],
    "Songwriting": ["https://www.songwriting.net", "https://www.americansongwriter.com", "https://www.songwriteruniverse.com"],
    "Album Reviews": ["https://pitchfork.com/reviews/albums", "https://www.metacritic.com/music", "https://www.rollingstone.com/music/music-album-reviews"],
    "Music Technology": ["https://www.musictech.net", "https://www.gearnews.com", "https://www.synthtopia.com"],
    "Music Business": ["https://www.musicbusinessworldwide.com", "https://www.musicindustryhowto.com", "https://www.hypebot.com"],
    "Classical Music": ["https://www.classicfm.com", "https://www.bbc.co.uk/programmes/genres/music/classical", "https://www.gramophone.co.uk"],
    "Jazz & Blues": ["https://www.jazzwise.com", "https://www.allaboutjazz.com", "https://www.downbeat.com"],
    "Rock & Metal": ["https://www.loudersound.com", "https://www.kerrang.com", "https://www.metalinjection.net"],
    "Hip-Hop & Rap": ["https://www.xxlmag.com", "https://www.hotnewhiphop.com", "https://www.complex.com/music"],
    "Electronic Dance Music (EDM)": ["https://www.youredm.com", "https://www.edmtunes.com", "https://www.dancingastronaut.com"],
    "Film Scores & Soundtracks": ["https://www.soundtrack.net", "https://www.filmscoremonthly.com", "https://www.classicfm.com/discover-music/periods-genres/film-tv"],
    "World Music": ["https://www.worldmusic.net", "https://www.songlines.co.uk", "https://www.afropop.org"],
}

# Function to clean text using NLP techniques
def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_text = [word for word in words if word not in stop_words]
    
    return ' '.join(filtered_text)

# Function to store extracted text data into individual files
def store_text_file(categories):
    dataset_dir = "./music_text_dataset"
    os.makedirs(dataset_dir, exist_ok=True)

    for category, urls in categories.items():
        combined_text = ""
        for url in urls:
            text = extract_text(url)
            combined_text += text + "\n\n"

        # Clean the extracted text
        combined_text = clean_text(combined_text)

        # Save to file
        filename = f"{dataset_dir}/{category}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(combined_text)
        print(f"Saved: {filename}")

# Run the dataset collection process
store_text_file(categories)
```
